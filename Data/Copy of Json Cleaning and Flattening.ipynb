```python
import json 
import pandas as pd 
import numpy as np 
from pprint import pprint
```

```python
reviews = pd.read_json("yelp_academic_dataset_review.json", lines=True)
business=pd.read_json("yelp_academic_dataset_business.json", lines=True)
```

```python
reviews.head()
```

```python
business.info()
```

```python
business=business.dropna()
```

```python
business=business[business["categories"].str.contains("Restaurants")]
```

```python
business.groupby("city").count().sort_values("business_id")
```

```python
business=business[business["city"]=="Philadelphia"]
```

```python
df=reviews.merge(business, on="business_id", how="left").dropna()
```

```python
grouped_business_df=df.groupby("business_id").count()
```

```python
grouped_business_df
```

```python
business=grouped_business_df[grouped_business_df["review_id"]>=500].index
```

```python
business
```

```python
df=df[df["business_id"].isin(business)]
```

```python
grouped_users_df=df.groupby("user_id").count()
```

```python
users=grouped_users_df[grouped_users_df["business_id"]>=15].index
```

```python
df_test=df[df["user_id"].isin(users)]
```

```python
df_test.to_csv("business_reviews.csv")
```
